```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nedznicy.csv')
```

```python
plt.style.use('dark_background')
```

```python
display(df)
```

```python
g = nx.Graph()
```

```python
#looking for the greatest weight:
tmp = df['value'].tolist()
greatest_value = max(tmp)
```

```python
#drawing edges:
for i in range(len(df)):
    g.add_edge(df['source'][i], df['target'][i], weight = df['value'][i])
```

```python
sizes = [g.degree(x)*150 for x in list(g)]
```

```python
w = [9*x/greatest_value for x in list(nx.get_edge_attributes(g,'weight').values())]


fig, ax = plt.subplots(figsize = (30, 20))

nx.draw_networkx(g, label = True, 
                font_size = 10, 
                 font_weight = 'bold',
                 font_color = 'white',
                node_color = 'firebrick', 
                node_size = sizes,
                width = list(w),
                 edge_color="cornflowerblue", 
                pos = nx.kamada_kawai_layout(g))

plt.title('Social graph of characters from book "NÄ™dznicy"',  fontsize = 30)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)



#plt.savefig('nedznicy_outcome.pdf')
```
