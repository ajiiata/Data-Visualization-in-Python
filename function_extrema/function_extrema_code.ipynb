```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 120000 #number of points

x = np.linspace(0, 2, n+1) #domain
y = 8*x**2-3*x**3-4*x #f(x) equation 

fig = plt.figure() 
ax = fig.add_subplot() 

ax.plot(x, y, color = 'blue', linewidth = 2) #function color and width


dydx = np.gradient(y, x) #we are looking for extrema
dydx[1:] * dydx[:-1]<0

ekstrema_x = x[1:][dydx[1:]*dydx[:-1]<0] 
ekstrema_y = y[1:][dydx[1:]*dydx[:-1]<0] 

ax.scatter(ekstrema_x, ekstrema_y, #plotting points
           color = "black",
           marker = 'o',
           zorder = 3 )

ax.spines['left'].set_position('center') #centering OY
ax.spines['bottom'].set_position('center')#centering OX

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_ylabel("y", loc="top") #podpisuje oś y 
ax.set_xlabel("x", loc="right") #podpisuje oś x 


ax.set_xlim(0, 2) #ustawiam limicik od 0 do 2

ax.set_xticks(ekstrema_x) 
ax.set_yticks(ekstrema_y) 

plt.grid(True) #making grid visible

ax.set_facecolor('moccasin') #bg color

plt.text(0.20, 1,'$f(x)=  8 \cdot x^2 - 3 \cdot x^3-4\cdot x$', horizontalalignment='left') #adding function equation on the plot

plt.title("$f(x)$ extrema") #plot title

plt.savefig('function_extrema.png') #saving picture

plt.show() #displaying plot
```
